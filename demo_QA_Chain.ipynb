```python
from dotenv import load_dotenv
import openai
import os
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import CSVLoader
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import DocArrayInMemorySearch
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from IPython.display import display, Markdown

load_dotenv()

openai.api_key = os.environ['OPENAI_API_KEY']

file = "./pdftext.csv"
loader = CSVLoader(file_path=file)
docs = loader.load()

 
text_splitter = RecursiveCharacterTextSplitter(
                chunk_size = 1000,
                chunk_overlap  = 0,
            )

splitted_documents = text_splitter.split_documents(docs)
splitted_documents
```

```python
embeddings = OpenAIEmbeddings()
docsearch = DocArrayInMemorySearch.from_documents(splitted_documents, embeddings)
```

```python
query = "서울 1박2일 겨울 여행코스 추천해줘"
docs = docsearch.similarity_search(query)
docs
```

```python
retriever = docsearch.as_retriever()
llm = ChatOpenAI(model_name='gpt-4-turbo-preview', temperature=0)

qa_stuff = RetrievalQA.from_chain_type(
    llm=llm, 
    chain_type="stuff", 
    retriever=retriever, 
    verbose=True
)

query = "Please list all 1 night and 2 days travel couses in markdown and summarize each one."
response = qa_stuff.invoke(query)
response
```

```python

from dotenv import load_dotenv
import openai
import os
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import CSVLoader
from langchain_openai import OpenAIEmbeddings
from langchain_openai import ChatOpenAI
from langchain_community.vectorstores import Chroma
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableParallel, RunnablePassthrough

load_dotenv()

openai.api_key = os.environ['OPENAI_API_KEY']

file = "/Users/jiapannan/PycharmProjects/pythonProject/pythonProject/1000 Assistant API/LangChain/rag/blog.csv"


loader = CSVLoader(file_path=file)
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
                chunk_size = 1000,
                chunk_overlap  = 0,
            )

splitted_documents = text_splitter.split_documents(docs)
embeddings = OpenAIEmbeddings()
docsearch = Chroma.from_documents(splitted_documents, embeddings)
retriever = docsearch.as_retriever()

template = """
Answer the question based only on the following context:
{context}

Question: {question}
"""

prompt = ChatPromptTemplate.from_template(template)
llm = ChatOpenAI(model_name='gpt-4-turbo-preview', temperature=0)
output_parser = StrOutputParser()
setup_and_retrieval = RunnableParallel(
    {"context": retriever, "question": RunnablePassthrough()}
)

chain = setup_and_retrieval | prompt | llm | output_parser

response = chain.invoke("서울 1박2일 겨울 여행코스 추천해줘")
```
